```python
import pandas as pd
import numpy as np
import re

```

# acs data

```python
dfcounty = pd.read_csv('acs2017_census_tract_data.csv', usecols=['TractId','State','County','TotalPop',
                                                                'Poverty','IncomePerCap'])
dfcounty.rename(columns = ({'TractId': 'ID', 'County':'Name', 'TotalPop': 'Population',
                     'IncomePerCap':'PerCapitaIncome'}), inplace=True)
dfcounty = dfcounty.dropna(axis=0)
```

```python
temp = []
for key, value in dfcounty['Name'].items():
    if "County" in value:
        temp_new = re.sub(" County", "", value)
    else:
        temp_new = value
    temp.append(temp_new)
    
dfcounty['Name'] = temp
```

```python
dfcounty['CountyAndState'] = dfcounty['Name']+', '+dfcounty['State']
dfcounty_temp = dfcounty['CountyAndState'].unique()
print(dfcounty_temp)
```

```python
len(dfcounty_temp)#3220
list1 = []
i = 1
for values in dfcounty_temp:
    list1.append(values)
    i+=1
list1
```

```python
len(list1)
len(dfcounty_temp)
```

```python
ID = []
Poverty_value = []
cnt = 1
for i, row in dfcounty.iterrows():
        if row.Name+', '+row.State in dfcounty_temp:
            ID.append(list1.index(row.Name+', '+row.State)+1)
            Poverty_value.append((row.Poverty/100)*row.Population)
```

```python
dfcounty['ID'] = ID
dfcounty['Poverty_val'] = Poverty_value
```

# A. County info

```python
dfcounty_info = dfcounty.groupby(['Name','State','ID'])
new_info = dfcounty_info.sum()
new_info = new_info.reset_index()
new_info
```

```python
new_info['Poverty_percent'] = (new_info['Poverty_val'] / new_info['Population']) * 100
new_info
```

```python
new_info[(new_info['Name'] == 'Loudoun') & (new_info['State'] == 'Virginia')] 
new_info[(new_info['Name'] == 'Washington') & (new_info['State'] == 'Oregon')]
new_info[(new_info['Name'] == 'Harlan') & (new_info['State'] == 'Kentucky')]
new_info[(new_info['Name'] == 'Malheur') & (new_info['State'] == 'Oregon')]

new_info[new_info['Population'] == new_info['Population'].max()]
new_info[new_info['Population'] == new_info['Population'].min()]
```

```python
county_info = new_info[["ID","Population","PerCapitaIncome","Poverty","Poverty_val"]]
```

# covid data

```python
dfcovid = pd.read_csv('COVID_county_data.csv')
dfcovid['Month'] = pd.DatetimeIndex(dfcovid['date']).month
dfcovid['Year'] = pd.DatetimeIndex(dfcovid['date']).year
dfcovid.rename(columns=({'county':'County','state':'State','cases':'Cases', 'deaths':'Deaths'}), inplace=True)
del dfcovid['fips']

dfcovid = dfcovid.dropna(axis=0)
```

```python
ID1 = []
cnt = 1
for i, row in dfcovid.iterrows():
        if row.County+', '+row.State in list1:
            ID1.append(list1.index(row.County+', '+row.State)+1)
        else: 
            ID1.append(9999)

dfcovid['ID'] = ID1
```

# B. Transform the COVID Data

```python
dfcovid_info = dfcovid.groupby(['Month','Year','County','State','ID'])
dfcovid_info
new_dfcovid = dfcovid_info.sum()
new_dfcovid = new_dfcovid.reset_index()
new_dfcovid
```

```python
new_dfcovid[(new_dfcovid['County'] == 'Malheur')
            & (new_dfcovid['State'] == 'Oregon')
            & (new_dfcovid['Month'] == 2) 
            & (new_dfcovid['Year'] == 2021)] 

new_dfcovid[(new_dfcovid['County'] == 'Malheur')
            & (new_dfcovid['State'] == 'Oregon')
            & (new_dfcovid['Month'] == 8) 
            & (new_dfcovid['Year'] == 2020)] 

new_dfcovid[(new_dfcovid['County'] == 'Malheur')
            & (new_dfcovid['State'] == 'Oregon')
            & (new_dfcovid['Month'] == 1) 
            & (new_dfcovid['Year'] == 2021)] 
```

```python
covid_info = dfcovid.groupby(['ID']
               ).agg({'Cases':sum, 'Deaths':sum})
covid_info = covid_info.reset_index()
covid_info
```

# C. Integrate COVID Data with ACS Data

```python
county_info = new_info[["ID","Name","State","Population","PerCapitaIncome","Poverty_val"]]
county_info
```

```python
covid_summary = pd.merge(county_info, covid_info, on = ["ID"])
```

```python
covid_summary
```

```python
covid_summary['PopulationPer100K'] = covid_summary['Population']/100000
covid_summary['TotalCasesPer100K'] = covid_summary['Cases']/covid_summary['PopulationPer100K']
covid_summary['TotalDeathsPer100K'] = covid_summary['Deaths']/covid_summary['PopulationPer100K']
covid_summary

covid_summary['Poverty_percent'] = (covid_summary['Poverty_val'] / covid_summary['Population']) * 100
covid_summary

```
